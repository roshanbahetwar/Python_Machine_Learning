## Data Balancing :- 

                            1. OverSampling (Upsampling)
                            2. UnderSampling (DownSampling)
                            
### 1. OverSampling(UpSampling)

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Roshan B\Desktop\ML\Dataset\creditcard.csv")
```

```python
df
```

```python
df['Class'].value_counts()
```

```python
x = df.drop('Class',axis=1)     # independent feature
```

```python
x
```

```python
y = df['Class']             # dependent feature
```

```python
y
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
obj = RandomOverSampler(sampling_strategy=0.70)
```

```python
obj
```

```python
x_oversample,y_oversample = obj.fit_resample(x,y)
```

```python
x_oversample
```

```python
y_oversample
```

```python
y_oversample.value_counts()
```

## 2. UnderSampling (DownSampling)

```python
from imblearn.under_sampling import NearMiss
```

```python
obj = NearMiss(sampling_strategy=0.70)
```

```python
x_undersample,y_undersample = obj.fit_resample(x,y)
```

```python
x_undersample
```

```python
y_undersample
```

```python
y_undersample.value_counts()
```
