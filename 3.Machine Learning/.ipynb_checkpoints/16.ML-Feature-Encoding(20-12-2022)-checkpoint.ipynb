### Encoding:-

two techniques of encoding:-

                        1. get_dummy (pandas)
                        2. one hot encoding (scikit learn)

## Method:- 1. get_dummy

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Roshan B\Desktop\ML\Dataset\Churn Modeling.csv")
```

```python
df
```

```python
df['encoded_gen'] = df['Gender'].map({'Female':1,'Male':0})
```

```python
df
```

```python
df1 = pd.get_dummies(df.Gender)
```

```python
df1
```

```python
df2 = pd.concat([df,df1],axis=1)
```

```python
df2
```

## Method:- 2. one_hot_encoding (nominal)

```python
pd.get_dummies(df.Geography)   # check by get_dummy method
```

```python
from sklearn.preprocessing import OneHotEncoder    # OneHotEncoder used
```

```python
df34 = OneHotEncoder(drop = 'first',sparse = False).fit_transform(df[['Geography']])     # output in array matrix
```

```python
df34          # output in array matrix
```

```python
df55 = pd.DataFrame(df34,columns=['Germany','Spain'])     # convert array matrix into dataframe
```

```python
df55
```

```python
final = pd.concat([df,df55],axis=1)             # concat df55 into df
```

```python
final
```

```python
final.drop(['Geography'],axis = 1,inplace= True)    # drop feature Geography after encoding
```

```python
final.drop(['Gender'],axis = 1,inplace= True)        # drop feature Gender after encoding
```

```python
final
```

## one_hot_encoding (ordinal)

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Roshan B\Desktop\ML\Dataset\bengaluru_house_prices.csv")
```

```python
df
```

```python
df['location'].value_counts()
```

```python
len(df['location'].value_counts())
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df['new_location'] = OrdinalEncoder().fit_transform(df['location'].values.reshape(-1,1))
```

```python
df
```

```python
df[df['location'] == 'Whitefield']
```

```python
df[df['location'] == 'Abshot Layout']
```

```python
df['society'].fillna(df['society'].mode()[0],inplace = True)
```
