## Simple Linear regression Algorithm

Predict canada's per capita income in year 2020. Using regression model and predict the per capita income fo canadian citizens in year 2020


```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df['year'],df['per capita income (US$)'], color='red',marker='*')
```

```python
new_df = df.drop('per capita income (US$)',axis= 'columns')
new_df
```

```python
model = linear_model.LinearRegression()
```

```python
model
```

```python
model.fit(new_df,df['per capita income (US$)'])    # model train
```

```python
model.predict([[2020]])  # predict the per capita income in 2020
```
