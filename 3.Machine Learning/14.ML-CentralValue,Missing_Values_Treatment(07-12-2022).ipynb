```python
# fill null value by '0' with new column 'new_deaths_0'

df['new_deaths_0'] = df['new_deaths'].fillna(0)
```

```python
# check null values
df.isnull().sum()
```

```python
# mean()

# fill null value by 'mean value of new_deaths' with new column 'new_deaths_mean' at last

df['new_deaths_mean'] = df['new_deaths'].fillna(df['new_deaths'].mean())
```

```python
df
```

```python
# median()

# fill null value by 'median value of new_deaths' with new column 'new_deaths_median' at last

df['new_deaths_median'] = df['new_deaths'].fillna(df['new_deaths'].median())
```

```python
df
```

```python
df['continent'].mode()      # continent column is categorical value thats why use mode()
```

```python
# mode()

# fill null value by 'mode value of continent' with new column 'continent_mode' at last

df['continent_mode'] = df['continent'].fillna(df['continent'].mode()[0])
```
