IQR:- Inter Quantile Range

IQR = q3-q1

### Outlier Detection & Removal

# # Outlier Detection & Removal-------> Method-1( Through IQR )

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Roshan B\Desktop\ML\Dataset\Churn Modeling.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
'''
for calculate IQR and q1,q3

IQR = q3-q1

'''
```

```python
q1 = df.Age.quantile(0.25)      # calculate q1
```

```python
q1
```

```python
q3 = df.Age.quantile(0.75)     # calculate q3
```

```python
q3
```

```python
IQR = q3-q1                    # calculate IQR
```

```python
IQR
```

```python
# calculate Lower and Upper Bound
```

```python
lower_bound = q1-(1.5*IQR)     # calculate lower bound
```

```python
lower_bound
```

```python
upper_bound = q3+(1.5*IQR)     # calculate upper bound
```

```python
upper_bound
```

```python
# Age without outlier

df[((df['Age']>= lower_bound) & (df['Age']<=upper_bound))]
```

```python
# Age with outlier ( negation symbol used for that):- this is outlier records

df[~((df['Age']>= lower_bound) & (df['Age']<=upper_bound))]
```

```python
df_outlier = df[~((df['Age']>= lower_bound) & (df['Age']<=upper_bound))]

df_outlier        # otlier in Age column (feature)
```

```python
'''
Sometimes due to outliers, data loss and due to that data removal is not feasible, so 
the option is Imputation to putting some values inplace of outliers
'''
```

```python
df_without_outlier = df[((df['Age']>= lower_bound) & (df['Age']<=upper_bound))]
```

```python
df_without_outlier['Age'].mean()     # check which value is most repeated for replace in outlier value by mean()
```

```python
df_without_outlier['Age'].median()   # check which value is most repeated for replace in outlier value by median()
```

```python
df_without_outlier['Age'].mode()     # check which value is most repeated for replace in outlier value  by mode()
```

```python
# now replacing median() value inplace outlier value
df_outlier['Age'] = df_without_outlier['Age'].median()  ''' 
                                                            replace median() value of 'Age' 
                                                    from df_without_outlier to df_outlier 'Age' after that concat to df1.
                                                                '''
```

```python
df_outlier
```

```python
df_without_outlier
```

```python
# concat 'df_without_outlier' and 'df_outlier'

df1 = pd.concat([df_without_outlier,df_outlier])
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.plot(df1['Age'])
```

```python
sns.pairplot(df1)
```

```python
sns.displot(df1['Age'])
```

```python
df1['Age'].value_counts()
```

#  Outlier Detection & Removal-------> Method-2( Through ZScore )

```python
df['zscore'] = (df['Age']-df['Age'].mean())/df.Age.std()
```

```python
df
```

```python
# -3 to 3    3std deviation
# -2 to 2    2std deviation
# -1 to 1    1std deviation
```

```python
df[(((df['zscore']>-3) & (df['zscore']<3)))]               # -3 to 3    3std deviation  without outlier
```

```python
df[~(((df['zscore']>-3) & (df['zscore']<3)))]               # -3 to 3    3std deviation  with outlier
```
