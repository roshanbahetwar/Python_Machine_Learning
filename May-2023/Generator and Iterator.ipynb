# Generator

```python
# in generator it generate a value one by one at time
# in Generator yield keyword use intead of return,its pause after result and give next value one by one
```

```python
def generator1():
    print('Hello Roshan')
    yield 17
    
    print('Hello Heena')
    yield 18
    
    print('Hello Ganesh')
    yield 19
```

```python
gen_val = generator1()
print(gen_val)               # it return objects, so access element next function to be used
```

```python
print(next(gen_val))
```

```python
print(next(gen_val))
```

```python
print(next(gen_val))
```

```python
print(next(gen_val))
```

```python
# using return

def generator2():
    print('Hello Roshan')
    yield 17
    
    return                      # here return used, after print it will stop, and result stop iteration error
    
    print('Hello Heena')
    yield 18
    
    print('Hello Ganesh')
    yield 19
```

```python
gen_val = generator2()
```

```python
print(next(gen_val))
```

```python
print(next(gen_val))
```

```python
# now using for loop    ----> it will not come stop iteration error


def generator3():
    print('Hello Roshan')
    yield 17
    
    print('Hello Heena')
    yield 18
    
    print('Hello Ganesh')
    yield 19
```

```python
gen_val = generator3()
```

```python
for i in gen_val:
    print(i)
```
