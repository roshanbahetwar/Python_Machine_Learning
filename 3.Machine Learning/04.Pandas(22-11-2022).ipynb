```python
# fetch record 'R' from column content_rating by using group by

df.groupby('content_rating').get_group('R')
```

```python
# fetch record 'PG' from column content_rating by using group by

df.groupby('content_rating').get_group('PG')
```

```python
# fetch record 'R' and 'Action' from column 'content_rating & genre' by using group by

df.groupby(['content_rating','genre']).get_group(('R','Action'))
```

```python
# fetch record 'PG' and 'Comedy' from column 'content_rating & genre'

df.groupby(['content_rating','genre']).get_group(('PG','Comedy'))
```

```python
# fetch record from same column as 'PG' and 'R'

df[(df['content_rating']=='PG') | (df['content_rating'] == 'R')]
```

```python
# fetch record from same column as 'PG' and 'R'

df[(df['content_rating'] == 'PG')  | (df['content_rating'] == 'R')]
```

```python
# fetch record from columns 'content_rating' & 'genre' of value R and Action

df.groupby(['content_rating','genre']).get_group(('R','Action'))
```

```python
df
```

```python
# fetch record from 'star_rating' of 9.3 & 7.4 from same column

df[(df['star_rating']== 9.3) | (df['star_rating']== 7.4)]
```

```python
# # fetch record from column content_rating as 'PG' and fetch column genre as 'Action'

df[(df['content_rating'] == 'PG') & (df['genre'] == 'Action')]
```

```python
# fetch record from column content_rating as 'PG' and fetch column genre as 'Action' by using group by

df.groupby(['content_rating','genre']).get_group(('PG','Action'))
```

```python
# Fetch max value from DatFrame

df[df['star_rating'] == df['star_rating'].max()]
```

```python
# Fetch min value from DatFrame

df[df['star_rating'] == df['star_rating'].min()]
```

```python
# fetch min & max value from DataFrame

df[(df['star_rating'] == df['star_rating'].max()) | (df['star_rating'] == df['star_rating'].min())]
```

```python
# min & and max count

df.groupby('genre').agg(['max','count'])
```
