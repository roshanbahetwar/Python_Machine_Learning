```python
# Merge

import pandas as pd
```

```python
df = pd.DataFrame({'id':
                  [101,102,103,104,105,106],
                  'name':['mohan','ganesh','roshan','joker','naresh','kaaml']
                  })
```

```python
df
```

```python
df1 = pd.DataFrame({'id':[105,104,102,110,145,103,124],
                   'city':['pune','mumbai','goa','nagpur','kasmir','delhi','punjab'],
                   'salary':[45165,41252,14256,22223,12322,66542,24577]
                   })
```

```python
df1
```

```python
df
```

```python
# use merge

pd.merge(df,df1,on = 'id')  # by default inner join
```

```python
pd.merge(df,df1,on = 'id')[['name','city']]  # fetch column only name & city
```

```python
pd.merge(df1,df, on = 'id')
```

```python
pd.merge(df,df1, on = 'id')[['id','name','salary']]
```

```python
# use indicator ---> to check from which dataframe

pd.merge(df,df1, on = 'id', indicator = True)
```

```python
# for left join

pd.merge(df,df1, on = 'id', how = 'left')
```

```python
# for right join

pd.merge(df,df1, on = 'id',how = 'right', indicator = True)
```

```python
# for outer join

pd.merge(df,df1, on = 'id', how = 'outer')
```

```python
pd.merge(df,df1, on = 'id', how = 'outer',indicator = True)
```
