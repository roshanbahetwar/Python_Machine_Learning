```python
import pandas as pd
```

```python
df = pd.DataFrame({'id':[101,102,103,104,105,106],
                  'name':['Roshan','Ganesh','Mohan','Amit','Kunal','Naresh']})
```

```python
df
```

```python
df1 = pd.DataFrame({'id':[101,103,106,600,650,700,800],
                  'city':['Pune','mumbai','Nagpur','Nashik','Aurangabad','Goa','Kolhapur'],
                  'salary':[4560,6581,9856,4587,6555,9654,6545]})
```

```python
df1
```

```python
pd.merge(df,df1,on='id', indicator = True)     # by default inner join (All Matched Records), # indicator = True ,means ,Check data came from which table
```

```python
pd.merge(df,df1,on = 'id',how = 'left', indicator = True)     # Left join ( All Left Records + Matched Records)
```

```python
pd.merge(df,df1,on = 'id',how = 'right',indicator = True)    # Right join ( All Right Records + Matched Records)
```

```python
pd.merge(df,df1,on = 'id', how = 'outer',indicator = True)    # full outer join (Matched+Unmatched Records)
```

```python
df2 = pd.DataFrame({'id':[101,102,103],
                   'name':['A','B','C'],
                   'city':['Pune','Mumbai','Satara']})
```

```python
df2
```

```python
df3 = pd.DataFrame({'id':[103,105,106,101,109,102],
                   'pincode':[442012,441207,441209,441015,441785,441254],
                   'District_Code':['G','M','R','V','Z','H']})
```

```python
df3
```

```python
df2
```

```python
pd.merge(df2,df3,on = 'id')
```

```python
pd.merge(df2,df3,on = 'id',how='left')[['id','city','pincode']]   # fetch perticular column with left join
```

```python
pd.merge(df2,df3, on = 'id',how = 'left',indicator = True)     # indicator = True ,means ,Check data came from which table
```
