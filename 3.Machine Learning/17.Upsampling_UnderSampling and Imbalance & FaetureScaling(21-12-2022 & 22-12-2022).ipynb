```python
# Note:- ValueError: could not convert string to float: '2100 - 2850'

# now need to convert string to float,

# to check which column has string, so column 'total_sqft' has string objects
```

```python
x.info()    # here 'total_sqft' column has in string(object) so its need to type convert into float64
```

```python
x['total_sqft'].to_list()    # check which value is in string
```

```python
def process(r):                                         # make a function to get numeric value by split
    if '-' in r:
        result = (float(r.split('-')[0]) + float(r.split('-')[1]))/2
        return result
    else:
        return r         
```

```python
x['new_total_sqft'] = x['total_sqft'].apply(process)
```

```python
import re               # import regular expression to fetch numeric value
```

```python
def fetch_number(r):             # make a function to fetch numeric value
    result = re.findall(r'\d+',str(r))
    return result[0]
```

```python
x['new_total_sqft'] = x['new_total_sqft'].apply(fetch_number)
```

```python
x.info()
```

```python
x['new_total_sqft'] = x['new_total_sqft'].astype('float')     # after fetching numeric value convert type object into float
```

```python
x.info()   # check type
```

```python
x.drop('total_sqft',axis = 1, inplace = True)    # now drop old column 'total_sqft' which have string object
```

```python
x
```

```python
# now use scale down
from sklearn.preprocessing import MinMaxScaler         # for MinMax Scaler
```

```python
obj = MinMaxScaler()
```

```python
obj.fit_transform(x)
```
