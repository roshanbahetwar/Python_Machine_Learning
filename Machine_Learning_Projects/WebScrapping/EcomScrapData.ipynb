```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.webscraper.io/test-sites/e-commerce/allinone/computers/laptops"

res = requests.get(url)
soup = BeautifulSoup(res.text,'lxml')   # lxml or html.parser, soup object created
print(soup)
```

```python
names = soup.find_all('a',class_ = 'title')

names_lst = []

for i in names:
    names_lst.append(i.text)
```

```python
print(names_lst)
```

```python
prices = soup.find_all('h4',class_ = 'float-end price card-title pull-right')

price_lst = []

for i in prices:
    price_lst.append(i.text)
```

```python
print(price_lst)
```

```python
reviews = soup.find_all('p',class_='float-end review-count')
reviews_lst = []

for i in reviews:
    reviews_lst.append(i.text)
```

```python
reviews_lst
```

```python
desc = soup.find_all('p',class_='description card-text')
desc_lst = []

for i in desc:
    desc_lst.append(i.text)
```

```python
desc_lst
```

```python
df = pd.DataFrame({'Laptop':names_lst,
                   'Price':price_lst,
                   'Reviews':reviews_lst,
                  'Decsription':desc_lst})
```

```python
df
```

```python
df.to_excel('laptop.xlxs')
```
