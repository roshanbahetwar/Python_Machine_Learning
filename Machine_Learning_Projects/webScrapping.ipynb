```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Establish connection with Server:
respond = requests.get('https://books.toscrape.com/catalogue/page-2.html')
```

```python
print(respond)
```

```python
soup = BeautifulSoup(respond.text,'html.parser')
```

```python
soup
```

```python
body = soup.body
```

```python
body
```

```python
# The Main Container
table=body.find_all('div',class_='col-sm-8 col-md-9')
```

```python
table
```

```python
# Extract Image URL
for i in table:
    y = i.find_all('div',class_ = 'image_container')
```

```python
y
```

```python
img_Tag=[]
for i in y:
    img_Tag.append(i.find_all('img',class_='thumbnail'))
```

```python
img_url = []
for i in range(int(len(img_Tag))):
    img_url.append(img_Tag[i][0]['src'])
```

```python
img_url
```

```python
for i in table:
    a_tag_name = i.find_all('h3')
```

```python
name=[]
for i in a_tag_name:
    name.append(i.a.text)
```

```python
name
```

```python
for i in table:
    price_tag=i.find_all('p',class_='price_color')
```

```python
price_tag
```

```python
price=[]
for i in price_tag:
    price.append(i.text)
```

```python
price
```

```python
df={"iamge_url":image_url,'name':name,'price':price}
data=pd.DataFrame(df)
data
```
