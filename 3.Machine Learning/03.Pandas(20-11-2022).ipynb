```python
# loc:-

# 1. loc :- this accessor selects rows and columns by labels

# ex. df.loc[row_labels,column_labels]
'''
df.loc['row1_label:row2_label','column1_label:column2_laabel']
'''
```

```python
# using loc for label-based selection

df.loc[:,'continent':'new_deaths']  # fetch all row and column from "continent" to "new_deaths"
```

```python
df.loc[1:50,:] # data fetch row from 1 to 50 and all columns
```

```python
df.loc[1:10,'continent':'new_deaths']  # data fetch from row 1 to 10 and column from "continent" to 'new_deaths'
```

```python
df.loc[[1,4,6],'continent':'new_cases']
```

```python
df.loc[[i for i in df.index if i%2==0],'continent':'new_cases']
```

```python
df.loc[[1,5,7,8],['continent','location','new_cases']]
```

```python
df.loc[[i for i in df.index if i%1000==0],['continent','location','total_deaths']]
```

```python
# iloc:-

# 1. iloc :- this accessor selects rows and columns by integer location.

# ex. df.iloc[row_position,column_position]
'''
df.iloc['row1_position:row2_position','column1_position:column2_position']
'''
```

```python
# using loc for index-based selection

df.iloc[[0,1,2,3],[3,4,5,6]]  # fetch row 0,1,2,3 and column index number 3,4,5,6
```

```python
df.iloc[1:10,0:5]  # fecth row from 1 to 9 and column from 0 to 4, (end is exlusive in iloc for row and column)
```
